```python
#SCATTER PLOT
#We use scatter plot to see the relationships between the 2 variables. We plot only a sample of 100 variables.

x = data['Age'].head(100)
y = data['Acceleration'].head(100)

plt.scatter(x,y)
plt.xlabel('Age')
plt.ylabel('Acceleration')

#Positioning
```

```python
len(data.Position.unique())
data.Position.unique()


defense_positions = ["LB", "CB", "RB", "RWB", "LWB"]
attack_positions = ["LW", "ST", "RW", "CF"]


for position in defense_positions:
    x = data[data["Position"]==position]["Age"]
    y = data[data["Position"]==position]["Acceleration"]
    color = 'blue'
    plt.scatter(x,y, color=color)
    
mean = data[data["Position"].isin(defense_positions)]["Acceleration"].mean()
plt.plot([x for x in range(50)],[mean for x in range(50)], color="blue")



for position in attack_positions:
    x = data[data["Position"]==position]["Age"]
    y = data[data["Position"]==position]["Acceleration"]
    color = 'red'
    plt.scatter(x,y, color=color)
    
    
mean = data[data["Position"].isin(attack_positions)]["Acceleration"].mean()
plt.plot([x for x in range(50)],[mean for x in range(50)], color="red")
```
