```python
#WEB SCRAPING
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://fr.wikipedia.org/wiki/Olympique_de_Marseille'
html = requests.get(url)
html
```

```python
soup = BeautifulSoup(html.text, 'lxml')
print(soup)
```

```python
#Display the names of the current players
players_table = soup.find_all('span', {'class': 'vcard'})
players_table

player_name = [span.find('a').get('title') for span in players_table if span.find('a') is not None]
player_name


```

```python
files = open("players.txt","w+")
```

```python
for elements in player_name:
    files.write(elements)
    files.write('\n')
files.close()
```
