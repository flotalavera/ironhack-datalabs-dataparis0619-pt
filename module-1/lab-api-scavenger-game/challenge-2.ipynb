```python
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
url ='https://api.github.com/repos/mayadesp/ironhack-datalabs-dataparis0619-pt/commits?since=2019-03-12&until=2019-07-19'

response = requests.get(url)
results = response.json()

flattened_data = json_normalize(results)
print(flattened_data)
```

```python
nb_commits = 0
for commits in results:
    nb_commits += 1
print(nb_commits)
```
