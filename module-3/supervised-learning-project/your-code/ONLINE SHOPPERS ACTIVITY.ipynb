```python
#Transform the categorical variables to numerical
online_shoppers = pd.get_dummies(online_shoppers,columns=['SpecialDay', 'Month', 'OperatingSystems','Browser','Region','TrafficType','VisitorType','Weekend'])
online_shoppers.columns
```

```python
#Transform the Revenue column to have 1 or 0 values instead boolean
#We use Label Encoder 

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

online_shoppers['Revenue'] = le.fit_transform(online_shoppers['Revenue'])
online_shoppers['Revenue'].value_counts()
```

```python
#SPLIT THE DATA
from sklearn.model_selection import train_test_split

y = online_shoppers.Revenue
x = online_shoppers.drop('Revenue', axis = 1)

X_train,X_test,y_train,y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
#ML ALGORITHM - CLASSIFICATION
```

```python
from sklearn.model_selection import cross_validate
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

cvscores = cross_val_score(LogisticRegression(), X_train, y_train, cv=5, verbose=10, scoring="f1")
cvscores.mean()
```

```python
cvscores = cross_val_score(LogisticRegression(), X_train, y_train, cv=5, verbose=10, scoring="accuracy")
cvscores.mean()
```

```python
from sklearn.ensemble import RandomForestClassifier
cvscores = cross_val_score(RandomForestClassifier(), X_train, y_train, cv=5, verbose=10, scoring="f1")
cvscores.mean()
```

```python
from sklearn.ensemble import RandomForestClassifier
cvscores = cross_val_score(RandomForestClassifier(), X_train, y_train, cv=5, verbose=10, scoring="accuracy")
cvscores.mean()
```

```python
from sklearn import svm
cvscores = cross_val_score(svm.SVC(), X_train, y_train, cv=5, verbose=10, scoring="f1")
cvscores.mean()
```

```python
cvscores = cross_val_score(svm.SVC(), X_train, y_train, cv=5, verbose=10, scoring="f1")
cvscores.mean()
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
plt.rcParams['figure.figsize'] = (8, 8)
sns.heatmap(cm ,annot = True)
```

```python
# train the model
print("Training Accuracy :", model.score(X_train, y_train))
print("Testing Accuracy :", model.score(X_test, y_test))
```

```python
print("Training Accuracy :", model.score(X_train, y_train))
print("Testing Accuracy :", model.score(X_test, y_test))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
print('Training set metrics:')
print('Accuracy:', accuracy_score(y_train, model.predict(X_train)))
print('Precision:', precision_score(y_train, model.predict(X_train)))
print('Recall:', recall_score(y_train, model.predict(X_train)))

print('Test set metrics:')
print('Accuracy:', accuracy_score(y_test, model.predict(X_test)))
print('Precision:', precision_score(y_test, model.predict(X_test)))
print('Recall:', recall_score(y_test, model.predict(X_test)))
```
